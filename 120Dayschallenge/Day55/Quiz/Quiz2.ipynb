# Quiz 2

1. What is the distance metric used in the KNN regressor?

   - Euclidean distance
   - Manhattan distance
   - Minkowski distance
   - All of the above

**Answer:** All of the above

**Explanation:** The KNN regressor can use various distance metrics, including Euclidean, Manhattan, and Minkowski distances, to measure the similarity between data points.

2. What is the optimal value of k in the KNN regressor?

   - A high value of k
   - A low value of k
   - It depends on the dataset
   - None of the above

**Answer:** It depends on the dataset

**Explanation:** The optimal value of k in the KNN regressor depends on factors such as the dataset's size, structure, and noise level. It is typically chosen using techniques like cross-validation.

3. Which of the following is not a disadvantage of the KNN regressor?

   - The need for feature scaling
   - The sensitivity to the number of features
   - The sensitivity to outliers
   - The high computation time

**Answer:** The high computation time

**Explanation:** Compared to some other algorithms, the KNN regressor generally has a lower computation time, especially during training.

4. How does the KNN regressor handle missing values?

   - It drops the data points with missing values
   - It replaces the missing values with the mean or median of the feature
   - It uses imputation methods to estimate the missing values
   - None of the above

**Answer:** It uses imputation methods to estimate the missing values

**Explanation:** The KNN regressor can use imputation methods like mean or median imputation or k-nearest neighbor imputation to handle missing values.

5. What is the main difference between the KNN classifier and the KNN regressor?

   - The KNN classifier predicts discrete values, while the KNN regressor predicts continuous values.
   - The KNN classifier uses a classification algorithm, while the KNN regressor uses a regression algorithm.
   - The KNN classifier uses a different distance metric than the KNN regressor.
   - None of the above

**Answer:** The KNN classifier predicts discrete values, while the KNN regressor predicts continuous values.

**Explanation:** The primary distinction is that the KNN classifier is used for classification tasks, while the KNN regressor is used for regression tasks.

6. In the KNN regressor, if k is set to 1 and there are 5 nearest neighbors, what value will be predicted for a new data point with the following distances to its neighbors: 1.2, 1.5, 1.8, 2.0, 2.5?

   - 1.5
   - 1.8
   - 2.0
   - 2.5

**Answer:** 1.5

**Explanation:** With k set to 1, the predicted value will be the value of the closest neighbor, which has a distance of 1.2 and a corresponding value of 1.5.

7. If the KNN regressor is trained on a dataset with 100 data points, and k is set to 10, how many neighbors will be used to predict the value of a new data point?

   - 5
   - 10
   - 20
   - 50

**Answer:** 10

**Explanation:** With k set to 10, the KNN regressor will utilize the 10 nearest neighbors to predict the value of a new data point.

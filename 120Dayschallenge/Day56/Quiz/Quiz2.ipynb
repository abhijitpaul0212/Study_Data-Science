# Quiz 2

**1. What is the mathematical technique used in PCA?**

   a) Linear Regression
   
   b) Singular Value Decomposition
   
   c) Gradient Descent
   
   d) Decision Trees

**Answer:** 
b) Singular Value Decomposition

**Explanation:** 
PCA is based on the mathematical technique of Singular Value Decomposition (SVD), which is used to decompose a matrix into its constituent parts. In the case of PCA, SVD is used to decompose the covariance matrix of a dataset.

**2. What is a Scree plot used for in PCA?**

   a) To visualize the eigenvalues of the principal components
   
   b) To visualize the correlation matrix of the dataset
   
   c) To visualize the original dataset
   
   d) To visualize the principal components

**Answer:** 
a) To visualize the eigenvalues of the principal components

**Explanation:** 
A Scree plot is a graphical representation of the eigenvalues of the principal components in a PCA analysis. It is used to determine the optimal number of principal components to retain in the dataset, based on the magnitude of the eigenvalues.

**3. Which of the following statements is true about PCA?**

   a) PCA can only be applied to datasets with two dimensions
   
   b) PCA is a supervised learning technique
   
   c) PCA is used to reduce the dimensionality of a dataset
   
   d) PCA is used to increase the dimensionality of a dataset

**Answer:** 
c) PCA is used to reduce the dimensionality of a dataset

**Explanation:** 
PCA is a technique used for reducing the dimensionality of a dataset by identifying the most important features that account for the greatest amount of variance in the data.

**4. What is the role of eigenvalues in PCA?**

   a) Eigenvalues are used to identify the most important features in the data
   
   b) Eigenvalues are used to create the principal components
   
   c) Eigenvalues are used to calculate the variance of the data
   
   d) Eigenvalues are used to determine the optimal number of principal components to retain

**Answer:** 
d) Eigenvalues are used to determine the optimal number of principal components to retain

**Explanation:** 
Eigenvalues are used to determine the amount of variance accounted for by each principal component. A Scree plot is used to visualize the eigenvalues and determine the optimal number of principal components to retain in the dataset.

**5. How is PCA useful in machine learning?**

   a) It helps in identifying the most important features for prediction
   
   b) It helps in creating a decision tree model
   
   c) It helps in classifying data into different categories
   
   d) It helps in evaluating the performance of a model

**Answer:** 
a) It helps in identifying the most important features for prediction

**Explanation:** 
PCA is useful in machine learning because it helps in identifying the most important features that account for the greatest amount of variance in the data.

**6. Which of the following is a limitation of PCA?**

   a) It is only applicable to datasets with a small number of features
   
   b) It requires prior knowledge of the dataset
   
   c) It may not always capture the important variations in the data
   
   d) It can only be used for supervised learning tasks

**Answer:** 
c) It may not always capture the important variations in the data

**Explanation:** 
One limitation of PCA is that it may not always capture the most important variations in the data, especially if the data has complex relationships between the features.

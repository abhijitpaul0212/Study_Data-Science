# Quiz 3

**1. What is the Eigen-Decomposition approach?**

   a) A method for calculating the inverse of a matrix
   
   b) A method for calculating the determinant of a matrix
   
   c) A method for diagonalizing a matrix
   
   d) A method for finding the trace of a matrix

**Answer:** 
c) A method for diagonalizing a matrix

**Explanation:** 
The Eigen-Decomposition approach is a method for finding a diagonal matrix that is similar to a given square matrix, through the use of its eigenvalues and eigenvectors.

**2. Which of the following conditions must be satisfied for a square matrix to be diagonalizable using the Eigen-Decomposition approach?**

   a) The matrix must be symmetric
   
   b) The matrix must have distinct eigenvalues
   
   c) The matrix must have a full rank
   
   d) The matrix must be invertible

**Answer:** 
b) The matrix must have distinct eigenvalues

**Explanation:** 
For a square matrix to be diagonalizable using the Eigen-Decomposition approach, it must have a complete set of linearly independent eigenvectors, which requires the matrix to have distinct eigenvalues.

**3. What is the significance of the spectral theorem in the context of the Eigen-Decomposition approach?**

   a) It provides a method for calculating the determinant of a matrix
   
   b) It guarantees that a square matrix is diagonalizable if it is symmetric
   
   c) It provides a way to calculate the eigenvalues of a matrix
   
   d) It ensures that the diagonalization of a matrix is unique

**Answer:** 
b) It guarantees that a square matrix is diagonalizable if it is symmetric

**Explanation:** 
The spectral theorem states that any real symmetric matrix can be diagonalized by an orthogonal matrix, and any complex Hermitian matrix can be diagonalized by a unitary matrix.

**4. How can the Eigen-Decomposition approach be extended to handle non-square matrices?**

   a) By using singular value decomposition
   
   b) By using principal component analysis
   
   c) By computing the Moore-Penrose pseudoinverse
   
   d) By using the QR decomposition

**Answer:** 
a) By using singular value decomposition

**Explanation:** 
Singular value decomposition (SVD) can be used to extend the Eigen-Decomposition approach to handle non-square matrices, by decomposing the matrix into three matrices: U, Σ, and V, where U and V are unitary matrices and Σ is a diagonal matrix.

**5. Which of the following is true of the eigenvectors of a matrix?**

   a) They are orthogonal to each other.
   
   b) They are the same as the eigenvalues.
   
   c) They are always real numbers.
   
   d) They have a determinant of 1.

**Answer:** 
a) They are orthogonal to each other.

**Explanation:** 
Eigenvectors associated with distinct eigenvalues of a symmetric or Hermitian matrix are always orthogonal to each other.

**6. What is the purpose of the orthogonal matrix in the Eigen-Decomposition approach?**

   a) To ensure that the eigenvalues are real numbers.
   
   b) To ensure that the eigenvalues are positive.
   
   c) To ensure that the eigenvectors are orthogonal to each other.
   
   d) To ensure that the matrix is diagonalizable.

**Answer:** 
c) To ensure that the eigenvectors are orthogonal to each other.

**Explanation:** 
An orthogonal matrix is used to diagonalize a symmetric or Hermitian matrix, and to ensure that the eigenvectors associated with distinct eigenvalues are orthogonal to each other.

**7. Which of the following methods can be used to find the eigenvalues of a matrix?**

   a) Gaussian elimination
   
   b) LU decomposition
   
   c) Power iteration method
   
   d) QR decomposition

**Answer:** 
c) Power iteration method

**Explanation:** 
The power iteration method is an iterative algorithm that can be used to find the eigenvalues of a matrix, by computing successive powers of the matrix multiplied by a random vector and normalizing the result.

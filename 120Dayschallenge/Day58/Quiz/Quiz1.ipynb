# Quiz 1

1. Which of the following statements is true about DBSCAN?

   - It requires the number of clusters as an input parameter.
   - It is a hierarchical clustering algorithm.
   - It can handle non-linearly separable clusters.
   - It assumes that clusters have a Gaussian distribution.

**Answer:** 
It can handle non-linearly separable clusters.

**Explanation:** 
DBSCAN stands for Density-Based Spatial Clustering of Applications with Noise. Unlike k-means, which requires the number of clusters as an input parameter, DBSCAN can automatically identify the number of clusters based on the density of data points. DBSCAN is a non-hierarchical clustering algorithm, and it can handle clusters of arbitrary shapes and sizes. Additionally, DBSCAN does not assume any particular distribution of data points within a cluster.

2. Which of the following input parameters of DBSCAN determines the minimum number of points that must be within a given radius to form a dense region?

   - epsilon
   - min_samples
   - max_samples
   - min_distance

**Answer:** 
min_samples

**Explanation:** 
DBSCAN operates by grouping data points into dense regions, which are defined as areas where there are a minimum number of points within a given radius (specified by the epsilon parameter). The min_samples parameter determines the minimum number of points that must be within this radius to form a dense region. Points that are not part of any dense region are considered noise.

   - Euclidean distance

**Answer:**

4. In DBSCAN, what is the meaning of the parameter epsilon?

   - The maximum number of points in a cluster.
   - The minimum number of points in a cluster.
   - The maximum distance between two points in a cluster.
   - The minimum distance between two points in a cluster.

**Answer:** 
The maximum distance between two points in a cluster.

**Explanation:** 
The epsilon parameter in DBSCAN specifies the maximum distance between two points for them to be considered part of the same cluster. Points that are within epsilon distance of each other are considered neighbors, and they are used to form dense regions.

   - DBSCAN is faster than k-means.
   - DBSCAN does not require any input parameters.

**Answer:** 
DBSCAN can handle non-linearly separable clusters.

**Explanation:** 
One of the main advantages of DBSCAN over k-means is its ability to handle non-linearly separable clusters. While k-means is based on the assumption that clusters are spherical and linearly separable, DBSCAN can handle clusters of arbitrary shapes and sizes.

7. Which of the following statements is true about DBSCAN?

   - It can only handle datasets with a low dimensionality.
   - It is sensitive to the choice of initial cluster centers.
   - It can be used to detect outliers as well as clusters.
   - It requires the input dataset to be preprocessed to remove noise.

**Answer:** 
It can be used to detect outliers as well as clusters.

**Explanation:** 
DBSCAN can be used not only for clustering, but also for outlier detection. Outliers are points that are not part of any dense region and are too far away from other points to be considered noise. This makes DBSCAN a versatile algorithm that can be used for a variety of data analysis tasks.
